```python
import pandas as pd
```

```python
all_data = pd.read_csv('../data/ene_keskeytykset_2005-2014.csv', parse_dates=['start'])
all_data
```

Update transformers where missing and change some data types to int

```python
all_data['transformers'] = all_data.transformers.fillna(1).astype(int)
all_data[['ur_ndp', 'ru_ndp']] = all_data.loc[:,['ur_ndp', 'ur_ndp']].fillna(-1).astype(int)
```

```python
all_data
```

Remove unnecessary columns and filter some rows which are for sure not caused by weather

```python
all_data.drop(columns=['Alkoi_vuosi','kk','pv','h','min','s'], inplace=True)
```

```python
filt_data = all_data[(~all_data.loc[:, 'cause'].isin(['R1','R2','ST1','ST2','ST3','ST4']))]
filt_data
```

```python
from sqlalchemy import create_engine
db_host = '***REMOVED***'
db_port = 5432
db_name = "postgres"
db_user = "***REMOVED***"
db_pass = "***REMOVED***"
def save_dataset(df, table_name='ene_outages'):
    """
    Save dataset into the db

    df : DataFrame
         DataFrame data
    """
    if df is None or len(df) < 1:
        return
    print('Storing classification set to db sasse.{}...'.format(table_name))

    # db_name, db_user, db_host, db_pass
    engine = create_engine('postgresql://{user}:{passwd}@{host}:5432/{db}'.format(user=db_user,
                                                                                 passwd=db_pass,
                                                                                 host=db_host,
                                                                                 db=db_name))

    index_name = 'id'
    df.to_sql(table_name, engine, schema='sasse', if_exists='append', index=False)

save_dataset(filt_data)
```
